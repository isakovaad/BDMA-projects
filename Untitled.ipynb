<h3>In the COMPAS dataset we have categorical variables like race, sex etc. So, they don't have an ordinal relationship. 
That is why we use One-Hot Encoding.</h3>

```python
# One-Hot Encoding
categorical_columns = ['race', 'sex', 'age_cat', 'is_recid', 'two_year_recid']
compas_data = pd.get_dummies(compas_data, columns=categorical_columns)
```

<h1>Reweighing for Bias Mitigation</h1>

```python
from aif360.algorithms.preprocessing import Reweighing
from aif360.datasets import StandardDataset
```

```python
# Assuming compas_data is your pandas DataFrame after preprocessing

# Convert your DataFrame to AIF360's StandardDataset format
# Note: Replace 'label_column', 'race', 'sex', and 'age_less_25' with the actual column names
compas_aif360_dataset = StandardDataset(compas_data, 
                                        label_name='label_column', 
                                        favorable_classes=[1], 
                                        protected_attribute_names=['race', 'sex', 'age_less_25'], 
                                        privileged_classes=[[1], [1], [1]])  # Adjust based on your data

# Applying reweighing for the 'race' attribute
rw = Reweighing(unprivileged_groups=[{'race': 0}], privileged_groups=[{'race': 1}])
compas_reweighed = rw.fit_transform(compas_aif360_dataset)

# You would repeat similar steps for 'sex' and 'age_less_25'
```
